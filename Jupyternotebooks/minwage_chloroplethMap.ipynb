```python
#nbi:hide_in
import plotly
import plotly.express as px
import pandas as pd

from textwrap import wrap 
named_colorscales = px.colors.named_colorscales()
# print("\n".join(wrap("".join('{:<12}'.format(c) for c in named_colorscales), 96)))

df = pd.read_csv(r'C:\Users\bxprd\Data Analytics Bootcamp\Git_Repos\data-analysis-project-2\Data\minwage_code.csv', encoding='cp1252')
df = df[(df['State']!= 'District of Columbia' )]
```

```python
#nbi:hide_in
df = df[(df['Year']>= 1970 )]
```
